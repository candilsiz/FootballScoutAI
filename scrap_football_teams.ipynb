```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import re
from io import StringIO
```

```python
leagues_team_dict = {}

def Soup(url):
    data = requests.get(url)
    soup = BeautifulSoup(data.text, 'lxml')
    return soup

def get_team_urls(url):
    soup = Soup(url)
    time.sleep(1)
    standings_table = None
    while not standings_table:
        standings_table = soup.select('table.stats_table')
        if standings_table:
            standings_table = standings_table[0]
        else:
            print(f"Standings table not found, retrying...")
            time.sleep(2)
            soup = Soup(url)
    links = standings_table.find_all('a')
    links = [link.get("href") for link in links]
    links = [link for link in links if '/squads/' in link]
    return [f'https://fbref.com{link}' for link in links]

league_urls = {
    "Premier League": "https://fbref.com/en/comps/9/Premier-League-Stats",
    "Bundesliga": "https://fbref.com/en/comps/20/Bundesliga-Stats",
    "La Liga": "https://fbref.com/en/comps/12/La-Liga-Stats",
    "Serie A": "https://fbref.com/en/comps/11/Serie-A-Stats",
    "Ligue 1": "https://fbref.com/en/comps/13/Ligue-1-Stats",
    "Eredivisie": "https://fbref.com/en/comps/23/Eredivisie-Stats",
    "Super Lig": "https://fbref.com/en/comps/26/Super-Lig-Stats"
}
for league, url in league_urls.items():
    team_urls = get_team_urls(url)
    leagues_team_dict[league] = team_urls
    print(league, 'Finished')
    
print(leagues_team_dict)
```

```python
def extract_team_name(url):
    match = re.search(r'/squads/[a-z0-9]+/([a-zA-Z0-9-]+)-Stats', url)
    if match:
        team_name = match.group(1)
        team_name = team_name.replace('-', ' ')
        return team_name
    else:
        return None
url = 'https://fbref.com/en/squads/0e72edf2/Hellas-Verona-Stats'
print('Team name: ', extract_team_name(url))
```

```python
for leagues, team_list in leagues_team_dict.items():
    for teams in team_list:
        print(teams)
```

```python
all_match_tables = list()
for league, team_list in leagues_team_dict.items():
    print('\nFetching', league, 'Table Data\n')
    for team_url in team_list:
        while True:
            try:
                data = requests.get(team_url)
                html_string = StringIO(data.text)
                time.sleep(1)
                scores_fixture = pd.read_html(html_string, match='Scores & Fixtures')
                if scores_fixture:
                    scores_fixture_df = scores_fixture[0]
                    scores_fixture_df['Team'] = extract_team_name(team_url)
                    all_match_tables.append(scores_fixture_df)
                break  
            except ValueError:
                print(f"No 'Scores & Fixtures' table found for URL: {team_url}, retrying...")
                time.sleep(1) 
            except Exception as e:
                print(f"An error occurred for URL: {team_url} - {e}")
                break  
```

```python
match_df = pd.concat(all_match_tables)
match_df
```
